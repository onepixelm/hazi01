```python
#!wget "http://www.wavsource.com/snds_2018-06-03_5106726768923853/people/politics/kennedy_moon.wav"

import matplotlib.pyplot as plot
from scipy.io import wavfile

samplingFrequency, signalData = wavfile.read('kennedy_moon.wav')

# Plot the signal read from wav file

plot.subplot(211)
plot.title('Spectrogram of a speech of Kennedy')
plot.plot(signalData)
plot.xlabel('Sample')
plot.ylabel('Amplitude')

plot.subplot(212)
plot.specgram(signalData,Fs=samplingFrequency)

plot.xlabel('Time')
plot.ylabel('Frequency')
plot.show()


```
